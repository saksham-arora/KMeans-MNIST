```python
# data_reader = DataReader('/Users/sakshamarora/ML-HW4-KMeans/')
# tr_data, tr_class_labels, tr_subclass_labels = data_reader.get_train_data() 
tr_data = [line.split() for line in open('mnist_raw_data.txt').readlines()]
tr_class_labels=[int(line.strip('\n')) for line in open('mnist_labels.txt','r').readlines()]
npa = np.array(tr_data, dtype=np.uint8)
kmeans = KMeans(n_clusters=5,max_iter=200)
kmeans.fit(npa,tr_class_labels)
kmeans.plot_imgs(npa,25,True)
# print(type(tr_class_labels))
# print(tr_class_labels)
# print(len(tr_class_labels))
```

```python
kmeans.plot_imgs(kmeans.centroids,len(kmeans.centroids))
```

```python
print('Loss')
plt.plot(range(kmeans.iterations),kmeans.loss_per_iteration)
plt.show()
```

```python
for key,data in list(kmeans.clusters['data'].items()):
    print('Cluster:',key,'Label:',kmeans.clusters_labels[key])
    kmeans.plot_imgs(data[:min(25,data.shape[0])],min(25,data.shape[0]))
```

```python
print('[cluster_label,no_occurrence_of_label,total_samples_in_cluster,cluster_accuracy]\n',kmeans.clusters_info)
print('Accuracy:',kmeans.accuracy)
```
