## Implementing KMeans on MNIST

```python
import os
from KMeans import KMeans
import matplotlib.pyplot as plt
import numpy as np
```

## Implement the k-means clustering algorithm. You will only use your algorithm for this dataset, so you can hard-wire in the number of instances and the size of each instance. The goal is not to write a generic version of the algorithm (though you can if you wish). The goal is to understand how it works on real data. You will need to try different values of k so that must be a parameter.

- Below is the code snippet of the implementation of KMeans algorithm. mnist_raw_data.txt and mnist_labels.txt are the text files which were used to train. mnist_raw_data.txt contains all the instances(i.e 10000 rows) and mnist_labels.txt contains all the corrected labels for those instances. I've used numpy library to compute the distance between centroids and when the differnece is close 0 the algorithm will converge. 

- Part2 below shows the iter_avg and incorrect_avg. iter_avg is the average number of iterations it took to converge where else incorrect_avg is the average number of incorrect labels in the cluseter.



```python
training_data = [line.split() for line in open('mnist_raw_data.txt').readlines()]
training_labels=[int(line.strip('\n')) for line in open('mnist_labels.txt','r').readlines()]
train_data= np.array(training_data, dtype=np.uint8)
kmeans = KMeans(n_clusters=10,max_iter=200)
iter_avg=[None for _ in range(10)]
incorrect_avg=[None for _ in range(10)]
#Part2 of the HW
for i in range(10):
    print("Runtime - ",i+1)
    iter_avg[i],incorrect_avg[i]=kmeans.form_cluster(train_data,training_labels)
print("Average number of iterations to converge : ",sum(iter_avg)//10)
print("Average number of incorrects labels : ",sum(incorrect_avg)//10)
kmeans.plot_imgs(train_data,25,True)
```

```python
print(sum(iter_avg)//10)
print(sum(incorrect_avg)//10)
kmeans.plot_imgs(kmeans.centroids,len(kmeans.centroids))
```

```python
for key,data in list(kmeans.clusters['data'].items()):
    print('Cluster:',key,'Label:',kmeans.clusters_labels[key])
    kmeans.plot_imgs(data[:min(25,data.shape[0])],min(25,data.shape[0]))
```

```python
incorrect_all=[0 for _ in range(10)]
for i in range(10):
    print("Cluster Label : ",kmeans.clusters_info[i][0])
    print("Number of Accurate Labels : ",kmeans.clusters_info[i][1])
    print("Total number of samples : ",kmeans.clusters_info[i][2])
    print("Incorrect Labels : ",kmeans.clusters_info[i][2] - kmeans.clusters_info[i][1])
    print("Cluster Accuracy : ",kmeans.clusters_info[i][3])
    print("---------------------------------------------")
    incorrect_all.append(kmeans.clusters_info[i][2] - kmeans.clusters_info[i][1])
print('Accuracy:',kmeans.accuracy)
print('Sum over all the Clusters:',sum(incorrect_all))
```
